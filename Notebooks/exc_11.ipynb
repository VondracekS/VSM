```python
import pandas as pd
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data_credit = pd.read_csv("../data/GermanCredit.csv", sep=";", decimal=",")
```

```python
data_credit.columns
```

```python
data_credit
```

```python
sns.regplot(data=data_credit, x="Class", y="A00Amount")
```

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
X, y = data_credit.drop(["Class"], axis=1), data_credit["Class"]
features = ["A00Amount100", "A01Duration"]
log_fit = logit.fit(X[features], y)
```

```python
features
```

```python
for feat, estim in zip(["intercept"] + features, [log_fit.intercept_, log_fit.coef_]):
    print(f"{feat}:,{estim}")
```
